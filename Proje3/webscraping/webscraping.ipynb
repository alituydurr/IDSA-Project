```python
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import re
import numpy as np
import time
```

```python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Safari/537.36"}) # Safari/537.36. Chrome/103.0.0.0
    soup = bts(result.text, "html.parser")
    return soup
```

```python
from requests_html import AsyncHTMLSession

async def get_data():
    asession = AsyncHTMLSession()
    url = "https://www.tradingview.com/symbols/BTCUSDT/"
    response = await asession.get(url)
    
    # JavaScript'i çalıştır
    await response.html.arender()
    
    # İlgili span'ı seç (doğru CSS seçici kullanın)
    price = response.html.find('last-JWoJqCpY js-symbol-last', first=True)
    if price:
        print(f"BTC/USDT Fiyatı: {price.text}")
    else:
        print("Fiyat bulunamadı.")

# Asenkron fonksiyonu çalıştır
import asyncio
asyncio.run(get_data())
```
