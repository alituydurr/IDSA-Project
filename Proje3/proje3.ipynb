```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Dosya yolunu ekleyip csv olarak pandas kÃ¼tÃ¼phanesi ile okudum. Tek sÃ¼tun olarak geldi veri
#Veriyi incelediÄŸimde sÃ¼tun isimlerinin ";" iÅŸareti ile ayrÄ±ldÄ±ÄŸÄ±nÄ± gÃ¶rdÃ¼m ve "sep= ";" konmutu ile sÃ¼tunlarÄ± dÃ¼zenledim
df = pd.read_csv("D:\ISTDSA_Proje\Solana_01.01.2018-31.10.2024_historical_data_coinmarketcap.csv", sep=";")
df
```

```python
#SÃ¼tun isimleri bÃ¼yÃ¼k kÃ¼Ã§Ã¼k harf olarak karÄ±ÅŸÄ±k verilmiÅŸ ileride sorun yaÅŸamamak iÃ§in sÃ¼tun isimlerini dÃ¼zenliyorum
df.columns = df.columns.str.lower()
df.columns = df.columns.str.strip()
df
```

```python
#sÃ¼tunlarÄ±n typye larÄ±nÄ± inceleyelim
df.dtypes
#time kolonlarÄ± obj larak gelmiÅŸ dÃ¼zeltilmesi gerekir

df['timeopen'] = pd.to_datetime(df['timeopen']).dt.tz_localize(None).dt.floor('s')
df['timeclose'] = pd.to_datetime(df['timeclose']).dt.tz_localize(None).dt.floor('s')
df['timehigh'] = pd.to_datetime(df['timehigh']).dt.tz_localize(None)
df['timelow'] = pd.to_datetime(df['timelow']).dt.tz_localize(None)
df

#timestamp kolonunda verinin alÄ±ndÄ±ÄŸÄ± zamanlar yazÄ±lÄ±yor iÅŸimize yaramaz silebiliriz
df.drop('timestamp', axis=1 , inplace=True)

```

```python
#name kolonu her index iÃ§in aynÄ± bu yÃ¼zden Ã§Ä±karabiliriz
df.drop('name', axis=1, inplace=True)
```

```python
# 3 farklÄ± coini aynÄ± dataframede birleÅŸtireceÄŸimiz iÃ§in timehigh ve timelow sÃ¼tunlarÄ±mÄ±z iÅŸimize yaramayacak bu sebeple o sÃ¼tunlarÄ± da siliyoruz
df.drop('timehigh', axis=1, inplace=True)
df.drop('timelow', axis=1, inplace=True)
```

```python
#marketcap deÄŸeri (fiyat*dolaÅŸÄ±mdaki arz) fiyat analizinde karÅŸÄ±laÅŸtÄ±rma yaparken iÅŸimize yaramadÄ±ÄŸÄ± iÃ§in o sÃ¼tunu da siliyoruz
df.drop('marketcap', axis=1, inplace=True)
```

```python
#3 farklÄ± dataframe'i birleÅŸtireceÄŸimiz iÃ§in timeopen ve timeclose sÃ¼tunlarÄ±nÄ± sabit olarak alacaÄŸÄ±z bu sebeple diÄŸer sÃ¼tunlarÄ±n isimlerini deÄŸiÅŸtirdik
df.rename(columns={'open':'sol_open', 'high':'sol_high', 'low':'sol_low','close':'sol_close','volume':'sol_vol'} , inplace=True)
df_sol=df
df_sol
```
