```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Dosya yolunu ekleyip csv olarak pandas kütüphanesi ile okudum. Tek sütun olarak geldi veri
#Veriyi incelediğimde sütun isimlerinin ";" işareti ile ayrıldığını gördüm ve "sep= ";" konmutu ile sütunları düzenledim
df = pd.read_csv("D:\ISTDSA_Proje\Solana_01.01.2018-31.10.2024_historical_data_coinmarketcap.csv", sep=";")
df
```

```python
#Sütun isimleri büyük küçük harf olarak karışık verilmiş ileride sorun yaşamamak için sütun isimlerini düzenliyorum
df.columns = df.columns.str.lower()
df.columns = df.columns.str.strip()
df
```

```python
#sütunların typye larını inceleyelim
df.dtypes
#time kolonları obj larak gelmiş düzeltilmesi gerekir

df['timeopen'] = pd.to_datetime(df['timeopen']).dt.tz_localize(None).dt.floor('s')
df['timeclose'] = pd.to_datetime(df['timeclose']).dt.tz_localize(None).dt.floor('s')
df['timehigh'] = pd.to_datetime(df['timehigh']).dt.tz_localize(None)
df['timelow'] = pd.to_datetime(df['timelow']).dt.tz_localize(None)
df

#timestamp kolonunda verinin alındığı zamanlar yazılıyor işimize yaramaz silebiliriz
df.drop('timestamp', axis=1 , inplace=True)

```

```python
#name kolonu her index için aynı bu yüzden çıkarabiliriz
df.drop('name', axis=1, inplace=True)
```

```python
# 3 farklı coini aynı dataframede birleştireceğimiz için timehigh ve timelow sütunlarımız işimize yaramayacak bu sebeple o sütunları da siliyoruz
df.drop('timehigh', axis=1, inplace=True)
df.drop('timelow', axis=1, inplace=True)
```

```python
#marketcap değeri (fiyat*dolaşımdaki arz) fiyat analizinde karşılaştırma yaparken işimize yaramadığı için o sütunu da siliyoruz
df.drop('marketcap', axis=1, inplace=True)
```

```python
#3 farklı dataframe'i birleştireceğimiz için timeopen ve timeclose sütunlarını sabit olarak alacağız bu sebeple diğer sütunların isimlerini değiştirdik
df.rename(columns={'open':'sol_open', 'high':'sol_high', 'low':'sol_low','close':'sol_close','volume':'sol_vol'} , inplace=True)
df_sol=df
df_sol
```

Yukarıdaki işlemlerin aynısını BTC ve ETH için de yapacağız bu sebeple tek bir sayfada toparlayalım

```python
#BTC
df_btc = pd.read_csv("D:\ISTDSA_Proje\Bitcoin_01.01.2018-31.10.2024_historical_data_coinmarketcap.csv", sep=";") #dosyayı oku
df_btc.columns = df_btc.columns.str.lower() #kolon başlıklarını küçük harfe çevir
df_btc.columns = df_btc.columns.str.strip() #kolon isimlerindeki boşlukları at
df_btc['timeopen'] = pd.to_datetime(df_btc['timeopen']).dt.tz_localize(None).dt.floor('s') #object değerleri date formatına çevir
df_btc['timeclose'] = pd.to_datetime(df_btc['timeclose']).dt.tz_localize(None).dt.floor('s') #object değerleri date formatına çevir
df_btc['timehigh'] = pd.to_datetime(df_btc['timehigh']).dt.tz_localize(None) #object değerleri date formatına çevir
df_btc['timelow'] = pd.to_datetime(df_btc['timelow']).dt.tz_localize(None) #object değerleri date formatına çevir
df_btc.drop('name', axis=1, inplace=True) #name kolonunu at
df_btc.drop('timestamp', axis=1 , inplace=True) #timstamp kolonunu at
df_btc.drop('timehigh', axis=1, inplace=True) #timehigh kolonunu at
df_btc.drop('timelow', axis=1, inplace=True) #timelow kolonunu at
df_btc.drop('marketcap', axis=1, inplace=True) #marketcap kolonunu at
df_btc.rename(columns={'open':'btc_open', 'high':'btc_high', 'low':'btc_low','close':'btc_close','volume':'btc_vol'}, inplace=True) #kolonları tekrar isimlendiriyoruz
df_btc
```

```python
#ETH
df_eth = pd.read_csv("D:\ISTDSA_Proje\Ethereum_01.01.2018-31.10.2024_historical_data_coinmarketcap.csv", sep=";") #dosyayı oku
df_eth.columns = df_eth.columns.str.lower() #kolon başlıklarını küçük harfe çevir
df_eth.columns = df_eth.columns.str.strip() #kolon isimlerindeki boşlukları at
df_eth['timeopen'] = pd.to_datetime(df_eth['timeopen']).dt.tz_localize(None).dt.floor('s') #object değerleri date formatına çevir
df_eth['timeclose'] = pd.to_datetime(df_eth['timeclose']).dt.tz_localize(None).dt.floor('s') #object değerleri date formatına çevir
df_eth['timehigh'] = pd.to_datetime(df_eth['timehigh']).dt.tz_localize(None) #object değerleri date formatına çevir
df_eth['timelow'] = pd.to_datetime(df_eth['timelow']).dt.tz_localize(None) #object değerleri date formatına çevir
df_eth.drop('name', axis=1, inplace=True) #name kolonunu at
df_eth.drop('timestamp', axis=1 , inplace=True) #timstamp kolonunu at
df_eth.drop('timehigh', axis=1, inplace=True) #timehigh kolonunu at
df_eth.drop('timelow', axis=1, inplace=True) #timelow kolonunu at
df_eth.drop('marketcap', axis=1, inplace=True) #marketcap kolonunu at
df_eth.rename(columns={'open':'eth_open', 'high':'eth_high', 'low':'eth_low','close':'eth_close','volume':'eth_vol'}, inplace=True) #kolonları tekrar isimlendiriyoruz
df_eth
```
